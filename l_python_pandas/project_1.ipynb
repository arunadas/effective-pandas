```python
# create total_discount 
transactions["total_discount"] = transactions["RETAIL_DISC"] + transactions["COUPON_DISC"]
```

```python
transactions.head()
```

```python
transactions["percentage_discount"] = transactions["total_discount"]/transactions["SALES_VALUE"].abs()
```

```python
transactions.head()
```

```python
conditions = [
    (transactions["percentage_discount"]>1),
    (transactions["percentage_discount"]<0)
]

choices = [1,0]

transactions["percentage_discount"] = np.select(conditions, choices)
```

```python
transactions.head()
```

```python
transactions.drop(
    [
        'RETAIL_DISC',
        'COUPON_DISC',
        'COUPON_MATCH_DISC'
    ],
    axis=1,
    inplace=True
)
```

```python
transactions.head()
```

```python
transactions.head()
```

```python
### transaction 
transactions = (
    transactions
    .assign(total_discount = transactions["RETAIL_DISC"] + transactions["COUPON_DISC"],
      percentage_discount = (lambda x: (x["total_discount"]/x["SALES_VALUE"]).abs()))
    .drop(['RETAIL_DISC','COUPON_DISC','COUPON_MATCH_DISC'],axis=1)      
            
)
```

```python
transactions.head()
```

```python
transactions["percentage_discount"] = (transactions["percentage_discount"]
                                       .where(transactions["percentage_discount"]<1,1.0)
                                       .where(transactions["percentage_discount"]>0,0) #false
                                      )
```

### Household Analysis 
- plot the distribution of total sales value purchased at the household level
- what are the top 10 households by quantity purchased ?
- what were the top 10 households by sales value?
- plot the total sales value for our top 10 households by value , ordered from hihest to lowest 

```python
## By store
(transactions
 .groupby(['store_nbr'], as_index=False)
 .agg(st_target_mean=('target_pct','mean'),
      st_bonus=('bonus_payable','sum'))
).sort_values('st_bonus', ascending=False).head()
```

```python
#
household_level = (transactions
 .groupby(["household_key"])
 .agg(total_sales=('SALES_VALUE','sum'))
)
```
